```python
import gensim, logging
from gensim.models import word2vec
```

```python
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s',level=logging.INFO)

sentences = [['I am a hero', 'sentence'], ['She is a teacher', 'sentence']] 

# train word2vec on the two sentences  
for i in [1,2,3,4]:
    for j in [0,1]:
        print(f"sg ={j}, window={i}")
        model = word2vec.Word2Vec(sentences, size=256, min_count=1, window=i, workers=4, sg=j)
        print(model)
        print(f"similarity: {model.similarity('I am a hero','She is a teacher'):.3f}\n")
# sg=0 表示COBW, sg=1 表示skip-gram
```
