```python
#import required libraries 
#import OpenCV library
import cv2
#import matplotlib library
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# cv2.cvtColor is an OpenCV function to convert images to different color spaces
def convertToRGB(img): 
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def detect_faces(f_cascade, colored_img, scaleFactor = 1.1,minNeighbors=5):
 #just making a copy of image passed, so that passed image is not changed 
 img_copy = colored_img.copy()          
 
 #convert the test image to gray image as opencv face detector expects gray images
 gray = cv2.cvtColor(img_copy, cv2.COLOR_BGR2GRAY)          
 
 #let's detect multiscale (some images may be closer to camera than others) images
 faces = f_cascade.detectMultiScale(gray, scaleFactor=scaleFactor, minNeighbors=minNeighbors);          
 
 #go over list of faces and draw them as rectangles on original colored img
 for (x, y, w, h) in faces:
      cv2.rectangle(img_copy, (x, y), (x+w, y+h), (0, 255, 0), 2)              
 
 return len(faces), img_copy
```

```python
#load cascade classifier training file for haarcascade
haar_face_cascade_alt = cv2.CascadeClassifier('D:/3_learn/9_MachineLearning/data/Part06/haarcascade_frontalface_alt.xml')

#load cascade classifier training file for lbpcascade
lbpcascade_frontalface = cv2.CascadeClassifier('D:/3_learn/9_MachineLearning/data/Part06/lbpcascade_frontalface.xml')
lbpcascade_frontalface_improved = cv2.CascadeClassifier('D:/3_learn/9_MachineLearning/data/Part06/lbpcascade_frontalface_improved.xml')
```

```python
# load image
test5 = cv2.imread("D:/3_learn/9_MachineLearning/data/Part06/test5.jpg")

# call our function to detect faces with harr_face_cascade_alt classifier
for i in [1.2, 1.3]:
    for j in [1,3,5]:
        n_Faces, faces_detected_img = detect_faces(haar_face_cascade_alt, test5, scaleFactor=i, minNeighbors=j)
        print(f"scaleFactor{i},minNeighbors={j},n_Faces={n_Faces}")
```

```python
n_Faces, faces_detected_img = detect_faces(haar_face_cascade_alt, test5, scaleFactor=1.3, minNeighbors=1)
plt.imshow(convertToRGB(faces_detected_img))
```
