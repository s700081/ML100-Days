# Feature engineering

## data preprocessing
### numerical data: 
* fill missing values
* remove outliers
* reduce skewness
* normalization

### categorical data:
* Label Encoding
* One Hot Encoding
* Mean Encoding (prone to over-fitting)
* Counting
* Feature Hash (for categorical data with too many categories)

### time data: 
* deciphering time data into single unit (year, month, day, hour, minute, second) 
* cyclization (weekday, nth week, season)
* time segment

## feature combination
### combining numerical feature with numerical feature
* +-*/, depending largely on domain knowledge

### combining numerical feature with categorical feature
* Group by Encoding

## feature selection
* Filter (with correlation coeeficient): set criteria and remove features accordingly
* Wrapper : add/ remove features gradually and observe the change of target function (e.g. loss function)
* Embedded: based on the results of model fitting and remove features with lower coeeficient
    * L1 (LASSO)
    * GDBT

## feature evaluation
* feature importance: tree models only
    * weight: number of splits under this feature
    * cover: objects covered with this feature
    * gain: relative contribution of the corresponding feature to the model (based on loss function)
* permutation importance: any model

## leaf encoding 
* for classification problems only

```python
# feature selection
# make a simple fitting with GradientBoostingClassifier and see feature importance
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier()
gbc.fit(train,trainLabels)
feats = pd.Series(data=gbc.feature_importances_, index=train.columns)
feats = feats.sort_values(ascending=False)
feats
```

```python
# original features
print(f'cross_val_score of original 40 features: {cross_val_score(gbc, train, trainLabels, cv=5).mean()}')
# important features
high_feature30 = list(feats[:30].index)
high_feature20 = list(feats[:20].index)
high_feature10 = list(feats[:10].index)
high_feature5 = list(feats[:5].index)
high_feature3 = list(feats[:3].index)
print(f'cross_val_score of 30 important features: {cross_val_score(gbc, train[high_feature30], trainLabels, cv=5).mean()}')
print(f'cross_val_score of 20 important features: {cross_val_score(gbc, train[high_feature20], trainLabels, cv=5).mean()}')
print(f'cross_val_score of 10 important features: {cross_val_score(gbc, train[high_feature10], trainLabels, cv=5).mean()}')
print(f'cross_val_score of 5 important features: {cross_val_score(gbc, train[high_feature5], trainLabels, cv=5).mean()}')
print(f'cross_val_score of 3 important features: {cross_val_score(gbc, train[high_feature3], trainLabels, cv=5).mean()}')
```

# Machine learning

```python
# define a function to draw confusion matrix plot
def plot_confusion_matrix(confmat, target_names=None, title=None):
    plt.matshow(confmat, cmap=plt.cm.Blues, alpha=0.7)
    for i in range(confmat.shape[0]):
        for j in range(confmat.shape[1]):
            plt.text(x=j,y=i, s=f'{confmat[i][j]}',va='center',ha='center')
    
    if target_names is not None:
        plt.xticks(np.arange(len(target_names)),target_names)
        plt.yticks(np.arange(len(target_names)),target_names)
    plt.xlabel('Predicted label')
    plt.ylabel('True label')
    plt.title(title)
    plt.grid('off')
    plt.tight_layout()
```

```python
# gradient boosting classifier
# split training/ testing data
from sklearn.model_selection import train_test_split
from sklearn import metrics
train_high_feature10 =train[high_feature10]

x_train, x_test, y_train, y_test = train_test_split(train_high_feature10, trainLabels, test_size=0.25, random_state=0)
# create model
gbc = GradientBoostingClassifier()

# fitting training data with the model
gbc.fit(x_train, y_train)

# 將測試資料丟入模型取得預測結果
y_pred = gbc.predict(x_test)

# 利用accuracy評估模型預測結果
print(f'accuracy using gradientboosting classifier with default hyperparameters: {metrics.accuracy_score(y_test, y_pred):.3f}')

# 繪製 confusion matrix
confmat = metrics.confusion_matrix(y_test, y_pred)
plot_confusion_matrix(confmat, title = "gradientboosting classifier with default hyperparameters")
```

```python
# predict the 
submission = pd.DataFrame(gbc.predict(test[high_feature10]))
print(submission.shape)
```
